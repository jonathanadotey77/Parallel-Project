```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#connect to dbs

# Connect to the SQLite database
conn = sqlite3.connect("test.db")
```

```python

#load all data
# Define the tables to load
tables = ["knapsack_timings", "investor_balances", "wins", "investors"]

# Load each table into a pandas DataFrame
dfs = {}
for table in tables:
    dfs[table] = pd.read_sql(f"SELECT * FROM {table}", conn)
```

```python
# Plot the data from each table using matplotlib
for table, df in dfs.items():
    plt.figure()
    plt.title(table)
    for col in df.columns:
        if col != "index":
            plt.plot(df[col])
    plt.legend(df.columns)
    plt.show()
```

```python
# Close the database connection
conn.close()
```
